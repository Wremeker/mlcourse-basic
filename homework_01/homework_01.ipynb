## Задание 1: Зеркальное отражение цифр

Пользователь вводит пятизначное число. Программа должна зеркально отразить три центральные цифры. Первая и последняя остаются на местах.

<br>

**Примеры:**
* `23456 -> 25436`
* `30789 -> 38709`

```python
user_input = input('Enter 5 digits number: ')

if (len(user_input) != 5 or not user_input.isdigit()):
    print('Input must be a number and consist of 5 digits')
    exit(1)

print(user_input.replace(user_input[1:-1], (user_input[1:-1])[::-1]))
```

## Задание 4: Римские числа

Пользователь вводит целое положительное число, программа должна вернуть строку в виде римского числа.

<br>

**Примеры:**
* `3   -> III`
* `15  -> XV`
* `234 -> CCXXXIV`

```python
user_input = input('Enter positive number: ')
number = int(user_input)

if number < 1 or number > 3999:
    print('Number must be between 1 and 3999')
    exit(1)

alphabet_dict = {
    1: 'I',
    2: 'II',
    3: 'III',
    4: 'IV',
    5: 'V',
    6: 'VI',
    7: 'VII',
    8: 'VIII',
    9: 'IX',
    10: 'X',
    20: 'XX',
    30: 'XXX',
    40: 'XL',
    50: 'L',
    60: 'LX',
    70: 'LXX',
    80: 'LXXX',
    90: 'XC',
    100: 'C',
    200: 'CC',
    300: 'CCC',
    400: 'CD',
    500: 'D',
    600: 'DC',
    700: 'DCC',
    800: 'DCCC',
    900: 'CM',
    1000: 'M',
    2000: 'MM',
    3000: 'MMM'
}

roman = ''
for index in range(len(user_input)):
    zeros_count = len(user_input) - index - 1
    element = int(user_input[index])
    
    if element == 0:
        continue
    
    full_number = element * (10 ** zeros_count)
    roman += alphabet_dict.get(full_number, '')

print(f'Roman number: {roman}')
```
