## Задание 1: Конвертер регистров

Написать функцию, которая будет переводить snake_case в PascalCase и наоборот. 

Функция должна сама определять - какой формат ей передали. Можно добавить ключевой аргумент, который будет принудительно возвращать один из форматов.

<br>

**Примеры:**
* `otus_course     -> OtusCourse`
* `PythonIsTheBest -> python_is_the_best`



```python
def convert(input: str):
    output = ''

    if '_' in input:
        output = input[0].upper()
        for index in range(1, len(input)):
            letter = input[index]
            if input[index - 1] == '_': 
                output += letter.upper()
            else:
                output += letter if letter != '_' else ''
    
    else:
        for key, letter in enumerate(input):
            if letter.upper() == letter:
                prefix = '_' if key != 0 else ''
                output += prefix + letter.lower()
            else:
                output += letter 
    
    return output
```

## Задание 2: Проверка валидности даты

Написать функцию проверяющую валидность введенной даты.

<br>

**Примеры:**
* `29.02.2000 -> True`
* `29.02.2001 -> False`
* `31.04.1962 -> False`



```python
def valid_date(date: str) -> bool:
    try:
        day, month, year = map(int, date.split('.'))
    except ValueError:
        return False
    
    if month < 1 or month > 12:
        return False
    
    if day < 1:
        return False
    
    # https://ru.wikipedia.org/wiki/%D0%92%D0%B8%D1%81%D0%BE%D0%BA%D0%BE%D1%81%D0%BD%D1%8B%D0%B9_%D0%B3%D0%BE%D0%B4
    is_leap = (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)
    
    days_in_month = {
        1: 31, 2: 29 if is_leap else 28, 3: 31, 4: 30,
        5: 31, 6: 30, 7: 31, 8: 31,
        9: 30, 10: 31, 11: 30, 12: 31
    }

    return day <= days_in_month[month]
```

## Задание 3: Проверка на простое число

Функция проверки на простое число. Простые числа – это такие числа, которые делятся на себя и на единицу.

<br>

**Примеры:**
* `17 -> True`
* `20 -> False`
* `23 -> True`

```python
# Важный момент в постановке задачи, простое число, это то которое делится само на себя и на единицу и больше ни на что. 
# У простого числа ровно два делителя: 1 и оно само.

def is_prime_number(num: int):
    if num < 2:
        return False
    
    # +1 Включаем чтобы в случае с 49, мы учили семерку [2,3,4,5,6,7]
    # num ** 0.5 - это корень числа, проверяем до него, остальные числа проверять нет смысла
    for divisor in range(2, int(num ** 0.5) + 1):
        if num % divisor == 0:
            return False
        
    return True
```

## Задание 4: Учет пользователей

Пользователь в бесконечном цикле вводит данные пользователей: имя, затем фамилию, возраст и ID. Ввод продолжается до тех пор, пока не будет введено пустое поле. 

Пользователи заносятся в словарь, где ключ это ID пользователя, а остальные данные записываются в виде кортежа. 

**Программа должна проверять:**
* имя и фамилия состоят только из символов и начинаются с большой буквы - если не с большой, то заменяет букву на большую;
* возраст должен быть числом от 18 до 60;
* ID - целое число, дополненное до 8 знаков незначащими нулями, ID должен быть уникальным.

**Дополнительно:** написать функцию, которая будет выводить полученный словарь в виде таблицы.

```python

import uuid

database = {} # [id пользователя]: (name, surname, age)

while True:
    user_input = input('Введите данныe пользвателя: ')

    if user_input == '':
        break

    name, surname, age, id = user_input.split(',')

    if not name.strip().isalpha() or not surname.strip().isalpha():
        print('Имя и фамилия должны состоять только из букв ')
        exit(1)

    if int(age) < 18 or int(age) > 60:
        print('Недопустимый возраст')
        exit(1)
        
    id = uuid.uuid4().hex[:8]

    while id not in database:
        id = uuid.uuid4().hex[:8]
        database[id] = (name.strip().capitalize(), surname.strip().capitalize(), int(age))

print('Id, Имя, Фамилия, Возраст')
for key, id in enumerate(database.keys()):
    name, surname, age = database[id]
    print(f'{id} {name} {surname} {age}')
```
